```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame named 'outbound' with 'date', 'A5_envelopes', and 'A4_envelopes' columns

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 7))

# Define the width of each bar
bar_width = 0.35

# Create x-values for the bars for the first set of data
x1 = range(len(outbound.date))

# Create x-values for the bars for the second set of data
x2 = [x + bar_width for x in x1]

# Plot the first set of bars
ax.bar(x1, outbound.A5_envelopes, width=bar_width, label='A5 Envelopes')

# Plot the second set of bars
ax.bar(x2, outbound.A4_envelopes, width=bar_width, label='A4 Envelopes')

# 3rd plot
ax.bar(x2, outbound.clear_tapes, width=bar_width, label='clear tapes')

#4th plot
ax.bar(x2, outbound.Plastic_Bags_Medium, width=bar_width, label='Plastic_Bags_Medium')

#5th plot
ax.bar(x2, outbound['50KGS_Suck'], width=bar_width, label='50KGS_Suck')

#6th plot
ax.bar(x2, outbound['carton_boxes_(medium)'], width=bar_width, label='carton_boxes_(medium)')

#7th plot
ax.bar(x2, outbound['clear_tapes'], width=bar_width, label='clear_tapes')

# Set the x-axis labels to be the dates
ax.set_xticks([x + bar_width / 3 for x in x1])
ax.set_xticklabels(outbound.date, rotation=45, ha='right')

# Add a legend
ax.legend()

# Set labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Quantity')
ax.set_title('A5 vs A4 Envelopes Over Time')

# Show the plot
plt.show()
```
